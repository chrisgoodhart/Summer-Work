### Chapter 16. Classes and Objects - The Basics

#### Exercise 1

Add a distanceFromPoint method that works similar to distanceFromOrigin except that it takes a Point as a parameter and computes the distance between that point and self.

```python
#Exercise 1

#imports the math module
import math

#defines a class Point with attributes iX and iY
class Point:
    
    #the constructor with 2 arguments, iX and iY
    def __init__(self, iX, iY):
        self.x = iX
        self.y = iY
    
    #method to return X of a point
    def getX(self):
        return self.x
    
    #method to return Y of a point
    def getY(self):
        return self.y
    
    #method to return the distance from the origin of a point
    def distanceFromOrigin(self):
        return (((self.x ** 2) + (self.y ** 2)) ** 0.5)
    
    #method with another point as an argument that returns the distance between the points
    def distanceFromPoint(self, point):
        difx = (self.x - point.getX())
        dify = (self.y - point.getY())
        return math.sqrt((difx ** 2) + (dify ** 2))
#samples of the methods
a = Point(3, 4)
b = Point(6, 7)
print(a.getX())
print(a.distanceFromOrigin())
print(b.distanceFromOrigin())
print(a.distanceFromPoint(b))
```

### Chapter 17. Classes and Objects - Digging a Little Deeper

#### Exercises 7 & 8

We can represent a rectangle by knowing three things: the location of its lower left corner, its width, and its height. Create a class definition for a Rectangle class using this idea. To create a Rectangle object at location (4,5) with width 6 and height 5, we would do the following:

r = Rectangle(Point(4, 5), 6, 5)

Add the following accessor methods to the Rectangle class: getWidth, getHeight, __str__.

```python
#Exercises 7 & 8

#imports the math module
import math

#defines the Point class
class Point:
    
    def __init__(self, iX, iY):
        self.x = iX
        self.y = iY
    
    def getX(self):
        return self.x
    
    def getY(self):
        return self.y
    
    def distanceFromOrigin(self):
        return (((self.x ** 2) + (self.y ** 2)) ** 0.5)
    
    def distanceFromPoint(self, point):
        difx = (self.x - point.getX())
        dify = (self.y - point.getY())
        return math.sqrt((difx ** 2) + (dify ** 2))
    
    def slope_from_origin(self):
        if self.x == 0:
            return None
        else:
            return self.y / self.x
        
    def __str__(self):
        return "x=" + str(self.x) + ", y=" + str(self.y)

#defines the Rectangle class
class Rectangle:
    
    #the constructor with 3 parameters: the lower left corner, the width, and the height
    def __init__(self, p, w, h):
        self.width = w
        self.height = h
        self.lower_left_corner = p
     
    #method to return the width
    def getWidth(self):
        return self.width
    
    #method to return the height
    def getHeight(self):
        return self.height
    
    #method to allow rectangles to be printed
    def __str__(self):
        return "w=" + str(self.width) + ", h=" + str(self.height) + ", p= " + str(self.lower_left_corner)

#sample of printing a rectangle
r = Rectangle(Point(4, 5), 6, 5)
print(r)
```

### Chapter 17. Classes and Objects - Digging a Little Deeper

#### Exercises 9 & 10

Add a method area to the Rectangle class that returns the area of any instance:

r = Rectangle(Point(0, 0), 10, 5)  
test(r.area(), 50)

Write a perimeter method in the Rectangle class so that we can find the perimeter of any rectangle instance:

r = Rectangle(Point(0, 0), 10, 5)  
test(r.perimeter(), 30)

```python
#Exercises 9 & 10

#imports the math module
import math

#defines the Point class
class Point:
    
    def __init__(self, iX, iY):
        self.x = iX
        self.y = iY
    
    def getX(self):
        return self.x
    
    def getY(self):
        return self.y
    
    def distanceFromOrigin(self):
        return (((self.x ** 2) + (self.y ** 2)) ** 0.5)
    
    def distanceFromPoint(self, point):
        difx = (self.x - point.getX())
        dify = (self.y - point.getY())
        return math.sqrt((difx ** 2) + (dify ** 2))
    
    def slope_from_origin(self):
        if self.x == 0:
            return None
        else:
            return self.y / self.x
        
    def __str__(self):
        return "x=" + str(self.x) + ", y=" + str(self.y)

#defines the Rectangle class
class Rectangle:
    
    def __init__(self, p, w, h):
        self.width = w
        self.height = h
        self.lower_left_corner = p
        
    def getWidth(self):
        return self.width
    
    def getHeight(self):
        return self.height
    
    def __str__(self):
        return "w=" + str(self.width) + ", h=" + str(self.height) + ", p= " + str(self.lower_left_corner)
    
    #returns the area of the rectangle
    def area(self):
        return self.height * self.width
    
    #returns the perimeter of the rectangle
    def perimeter(self):
        return 2 * self.height + 2 * self.width

#samples of new methods
r = Rectangle(Point(0, 0), 10, 5)
print(r.perimeter())
print(r.area())
```
