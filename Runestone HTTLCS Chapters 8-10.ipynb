#### Exercise 1

Add a print statement to Newton’s sqrt function that prints out better each time it is calculated. Call your modified function with 25 as an argument and record the results.

```python
#Exercise 1

#defines the function newtonSqrt that applies Newton's method for finding square roots
def newtonSqrt(n):
    app = 0.5 * n
    
    #prints the first approximation
    print ("Start ", app)
    better = 0.5 * (app + n / app)
    while better != app:
        app = better
        
        #each time a better approximation is calculated, print the better approximation
        print ("Better ", app)
        better = 0.5 * (app + n / app)
    
    #once the approximation is "close enough" to the better approximation, print the final approximation
    print ("Final ", better)

#samples
newtonSqrt(25)
newtonSqrt(2)
x = int(input("Input a number"))
newtonSqrt(x)
```

Sepia Tone images are those brownish colored images that may remind you of times past. The formula for creating a sepia tone is as follows:

newR = (R × 0.393 + G × 0.769 + B × 0.189)
newG = (R × 0.349 + G × 0.686 + B × 0.168)
newB = (R × 0.272 + G × 0.534 + B × 0.131)

Write a function to convert an image to sepia tone. Hint: Remember that rgb values must be integers between 0 and 255.

```python
#Exericse 10

#imports the image module
import image

#defines a function sepiatone which, given an image, returns the image with sepia tones
def sepiatone(picture):
    sepia = image.EmptyImage(picture.getWidth(), picture.getHeight())
    
    #collects the RGB values for each pixel of the input image
    for c in range (picture.getWidth()):
        for r in range (picture.getHeight()):
            p = picture.getPixel(c, r)
            oldred = p.getRed()
            oldgreen = p.getGreen()
            oldblue = p.getBlue()
            
            #calculates the sepia tones for each color
            newred = (oldred * 0.393 + oldgreen * 0.769 + oldblue * 0.189)
            newgreen = (oldred * 0.349 + oldgreen * 0.686 + oldblue * 0.168)
            newblue = (oldred * 0.272 + oldgreen + 0.534 + oldblue * 0.131)
            
            #checks if the new color values exceed 255
            if newred > 255:
                newred = 255
            if newgreen > 255:
                newgreen = 255
            if newblue > 255:
                newblue = 255
            
            #fills in the sepia image with the new pixels
            newpixel = image.Pixel(newred, newgreen, newblue)
            sepia.setPixel(c, r, newpixel)
    return sepia

#instantiates the window and the test image of user input
win = image.ImageWin()
testimage = image.Image(input("File name of image?"))

#draws the input image with sepia tones
sepiaimage = sepiatone(testimage)
sepiaimage.draw(win)

win.exitonclick()
```

#### Exercise 14

Here is a dragon curve. Use 90 degrees.:

FX 

X -> X+YF+ 

Y -> -FX-Y

```python
#Exercise 14

#imports the turtle module
import turtle

#defines a function createLSystem that will take a string and iterate the rules n times
def createLSystem(n, axiom):
    start = axiom
    end = ""
    for i in range(n):
        end = process(start)
        start = end
    
    return end

#defines a function process that will iterate the rules on each character of a given string
def process(inString):
    processed = ""
    for char in inString:
        processed = processed + rules(char)
    
    return processed

#defines a function rules that applies the rules to a given character
def rules(character):
    if character == 'X':
        return 'X+YF+'
    elif character == 'Y':
        return '-FX-Y'
    else:
        return character

#defines a function drawLSystem that will use a string as instructions to draw the L system
def drawLSystem(turtle, instructions, angle, distance):
    for cmd in instructions:
        if cmd == 'F':
            turtle.forward(distance)
        elif cmd == 'B':
            turtle.backward(distance)
        elif cmd == '+':
            turtle.right(angle)
        elif cmd == '-':
            turtle.left(angle)

#defines the main function
def main():
    #asks for user input about iterations and distance
    iter = int(input("How many iterations?"))
    dist = int(input("How much distance?"))
    
    #creates the string that will be used as instructions
    instruct = createLSystem(iter, 'FX')
    print(instruct)
    
    #instantiates the turtle and window
    t = turtle.Turtle()
    wn = turtle.Screen()
    
    #draws the L system
    t.speed(20)
    drawLSystem(t, instruct, 90, dist)
    
    wn.exitonclick()
    
main()
```

#### Exercise 21

Write a function called rot13 that uses the Caesar cipher to encrypt a message. The Caesar cipher works like a substitution cipher but each character is replaced by the character 13 characters to ‘its right’ in the alphabet. So for example the letter a becomes the letter n. If a letter is past the middle of the alphabet then the counting wraps around to the letter a again, so n becomes a, o becomes b and so on. Hint: Whenever you talk about things wrapping around its a good idea to think of modulo arithmetic.

```python
#Exercise 21

def rot13(mess):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    newmess = ''
    for ch in mess:
        if ch == ' ':
            newmess = newmess + ' '
        else:
            newindex = alphabet.index(ch) + 13
            if newindex >= 26:
                newmess = newmess + alphabet[newindex % 26]
            else:
                newmess = newmess + alphabet[newindex]
    return newmess

print(rot13('abcde'))
print(rot13('nopqr'))
print(rot13(rot13('since rot thirteen is symmetric you should see this message')))
```

#### Exercise 12

Count how many words occur in a list up to and including the first occurrence of the word “sam”.

```python
#Exercise 12

#defines a function count that counts words up to and including the first word 'sam'
#if the sentence given does not contain 'sam' then returns 0
def count(lst):
    words = lst.split(' ')
    x = 0
    for w in words:
        if w == 'sam' or w == 'Sam' or w == 'sam.' or w == 'Sam.':
            x = x + 1
            return x
        else:
            x = x + 1
    return 0

#samples
print(count('hi bob and sam'))
print(count('hi bob'))
print(count('my friends are bob, jack, john, michael, and sam'))

string = input("Give a sentence: ")
print(count(string))

```

Write a function replace(s, old, new) that replaces all occurences of old with new in a string s:

test(replace('Mississippi', 'i', 'I'), 'MIssIssIppI')

s = 'I love spom!  Spom is my favorite food.  Spom, spom, spom, yum!'

test(replace(s, 'om', 'am'),

       'I love spam!  Spam is my favorite food.  Spam, spam, spam, yum!')

test(replace(s, 'o', 'a'),

       'I lave spam!  Spam is my favarite faad.  Spam, spam, spam, yum!')
       
Hint: use the split and join methods.

```python
#Exercise 14

#defines a function replace to swap in characters in a string (only works if old and new are characters)
def replace(s, old, new):
    replaced = ''
    for i in range (len(s)):
        if s[i] == old:
            replaced = replaced + new
        else:
            replaced = replaced + s[i]
    return replaced

string = input("What is the string? ")
old = input("What do you want to swap out? ")
new = input("What do you want to swap in? ")
replace(string, old, new)
```
