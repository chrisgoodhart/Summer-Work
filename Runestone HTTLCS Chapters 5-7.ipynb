Repeat the above exercise (Use a for statement to print 10 random numbers) but this time print 10 random numbers between 25 and 35, inclusive.

```python
#Exercise 17

#Imports the random module
import random

#Runs the loop 10 times
x = 10
for i in range(x):
    #Stores a random number in the interval [25, 36) in "r"
    r = random.random()*11+25
    
    #If "r" is greater than 35, rerun the random number generator
    while (r>35):
        r = random.random()*11+25
        
    #Prints each number
    print(r)
```

### Chapter 6. Functions

#### Exercise #4

Draw this pretty pattern.

```python
#Exercise 4

#Imports necessary modules
import turtle
import math

#Defines a function "pattern" to draw five squares
def pattern(turtle, sqsize, sqnumber):
    turtle.stamp()
    
    #Draws the center radii
    drawradii(turtle, sqsize/2, sqnumber*4)
    turtle.up()
    
    #Draws a given number of squares, each one rotated
    for i in range(sqnumber):
        a = 45/sqnumber
        if i>0:
            turtle.left(a)
        turtle.left(a)
        x = (math.sqrt(2)*sqsize)/2
        turtle.forward(x)
        turtle.down()
        turtle.left(135)
        drawsquare(turtle, sqsize)
        turtle.up()
        turtle.right(135)
        turtle.forward(-x)

#Defines a function "drawsquare" to draw a square given the side length
def drawsquare(turtle, size):
    turtle.down()
    for i in range(4):
        turtle.forward(size)
        turtle.left(90)

#Defines a function "drawradii" to draw a given number of radii of given length around the starting point
def drawradii(turtle, size, number):
    turtle.down()
    for i in range(20):
        turtle.forward(size)
        turtle.forward(-size)
        turtle.left(360/number)

#Instantiates the window and sets its color
wn = turtle.Screen()
wn.bgcolor("lightgreen")

#Instantiates the turtle and sets its features
bob = turtle.Turtle()
bob.color("blue")
bob.shape("turtle")
bob.speed(20)

#Runs function "pattern" to draw 5 squares with side lengths 200
pattern(bob, 200, 5)
bob.ht()

wn.exitonclick()
```

The two spirals in this picture differ only by the turn angle. Draw both.

```python
#Exercise 5

#imports the turtle module
import turtle

#defines a function spiral that uses recursion to draw a spiral shape given a turn angle
def spiral(turtle, angle):
    x = 1
    for i in range(150):
        turtle.forward(x)
        turtle.left(angle)
        x = x + 2

#instantiates the window and sets its color
wn = turtle.Screen()
wn.bgcolor("lightgreen")

#instantiates a turtle and sets its color and speed
bob = turtle.Turtle()
bob.color("blue")
bob.speed(20)

#prompts the user for the turn angle
ang = int(input("What is the turn angle?"))
spiral(bob, ang)

wn.exitonclick()
```

#### Exercise #10

Write a function is_rightangled which, given the length of three sides of a triangle, will determine whether the triangle is right-angled. Assume that the third argument to the function is always the longest side. It will return True if the triangle is right-angled, or False otherwise.

Hint: floating point arithmetic is not always exactly accurate, so it is not safe to test floating point numbers for equality. If a good programmer wants to know whether x is equal or close enough to y, they would probably code it up as

if  abs(x - y) < 0.001:      # if x is approximately equal to y
    ...


```python
#Exercise 10

#the function assumes that the third argument, "c," is always the largest side length
#defines a function is_rightangled to test if a triangle is right-angled given 3 sides
def is_rightangled(a, b, c):
    hyp2 = c ** 2
    legs2 = a ** 2 + b ** 2
    
    #tests if the pythagorean theorem is true
    if abs(hyp2 - legs2) < 0.001:
        return True
    return False

#samples
print(is_rightangled(3, 4, 5))
print(is_rightangled(3, 5, 6))

```

#### Exercise #13

Implement the calculator for the date of Easter.

The following algorithm computes the date for Easter Sunday for any year between 1900 to 2099.

Ask the user to enter a year. Compute the following:

a = year % 19, 
b = year % 4, 
c = year % 7, 
d = (19 * a + 24) % 30, 
e = (2 * b + 4 * c + 6 * d + 5) % 7, 
dateofeaster = 22 + d + e

Special note: The algorithm can give a date in April. Also, if the year is one of four special years (1954, 1981, 2049, or 2076) then subtract 7 from the date.

Your program should print an error message if the user provides a date that is out of range.



```python
#Exercise 7

#defines a function doeaster that calculates the date of easter given a year
def doeaster(year):
    if year <= 2099 and year >= 1900:
        a = year % 19
        b = year % 4
        c = year % 7
        d = (19 * a + 24) % 30
        e = (2 * b + 4 * c + 6 * d + 5) % 7
        dateofeaster = 22 + d + e
        if year == 1954 or year == 1981 or year == 2049 or year == 2076:
            dateofeaster = dateofeaster - 7
        if dateofeaster > 31:
            print("Date of Easter: April, ", (dateofeaster - 31))
        else:
            print("Date of Easter: March, ", dateofeaster)
    else:
        print("Year is not in acceptable range.")

#prompts the user for a year to test
y = int(input("What year?"))
doeaster(y)
```
